# IMPORTING DATA

```python
#importing all necesarry libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing dataset
df=pd.read_csv('DataScience_salaries_2024.csv')
```

```python
df.head()
```

# Statistic Data Analysis 

```python
df.info()
```

```python
df = df.drop(['salary', 'salary_currency','employee_residence'], axis=1)
```

```python
df.isna().sum()
```

```python
df.drop_duplicates()
```

```python
df.describe()
```

# Data Visualization

```python
plt.figure(figsize=(12, 6))
avg_salary_by_job = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)
top_10_jobs = avg_salary_by_job.head(10)
sns.barplot(x=top_10_jobs.index, y=top_10_jobs.values, palette='viridis')
plt.title('Top 10 Jobs by Average Salary')
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_salary_by_year = df.groupby('work_year')['salary_in_usd'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_salary_by_year, x='work_year', y='salary_in_usd', marker='o', color='b')
plt.xticks(avg_salary_by_year['work_year'].astype(int))
plt.title('Average Salary Across Work Years')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python
experience_order = ['EN', 'MI', 'SE', 'EX']  
avg_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().reindex(experience_order)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_salary_by_experience.index, y=avg_salary_by_experience.values, palette='viridis')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='employment_type', palette='magma')
plt.title('Employment Type Distribution')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```

```python


avg_salary_by_company_size = df.groupby('company_size')['salary_in_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_salary_by_company_size.index, y=avg_salary_by_company_size.values, palette='viridis')
plt.title('Average Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python


avg_salary_by_employment_type = df.groupby('employment_type')['salary_in_usd'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_salary_by_employment_type.index, y=avg_salary_by_employment_type.values, palette='viridis')
plt.title('Average Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python
avg_salary_by_remote = df.groupby('remote_ratio')['salary_in_usd'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_salary_by_remote, x='remote_ratio', y='salary_in_usd', palette='viridis')
plt.title('Average Salary by Remote Ratio')
plt.xlabel('Remote Work Ratio (%)')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python

avg_salary_df = df.groupby(['work_year', 'experience_level', 'employment_type', 'job_title', 
                             'remote_ratio', 'company_size'])['salary_in_usd'].mean().reset_index()

corr = avg_salary_df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation of Average Salaries with Other Variables')
plt.show()
```
